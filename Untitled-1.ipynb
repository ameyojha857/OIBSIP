```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import numpy as np
```

```python

unemployment_df = pd.read_csv(r"C:\Users\Ameyo\OneDrive\Desktop\Oasis infobyte\Unemployemnt rate Analysis using python\Unemployment_in_India.csv")
location_df = pd.read_csv(r"C:\Users\Ameyo\OneDrive\Desktop\Oasis infobyte\Unemployemnt rate Analysis using python\Unemployment_Rate_upto_11_2020.csv")
```

```python
print(unemployment_df.info())
print(unemployment_df.head())

print(location_df.info())
print(location_df.head())
```

```python
merged_df = pd.merge(unemployment_df, location_df, on=' Estimated Unemployment Rate (%)', how='inner')
```

```python

print(merged_df.columns)
```

```python

merged_df = merged_df.dropna()
```

```python

print(merged_df.describe())
```

```python
avg_unemployment_by_region = merged_df.groupby('Region.1')[' Estimated Unemployment Rate (%)'].mean()
print(avg_unemployment_by_region)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(x='Area', y=' Estimated Unemployment Rate (%)', hue='Region.1', data=merged_df, ci=None)
bar_plot.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.title('Unemployment Rate % by Area')
plt.xlabel('Area')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(14, 8))
scatter_plot = sns.scatterplot(x=' Estimated Labour Participation Rate (%)_x', y=' Estimated Employed_x', hue='Area', style='Region.1', data=merged_df, s=100, alpha=0.7)
scatter_plot.legend(loc='upper right', bbox_to_anchor=(1.2, 1))
plt.yticks(merged_df[' Estimated Employed_x'].values, [f'{int(val/1000000)}K' for val in merged_df[' Estimated Employed_x'].values])
plt.title('Labour Participation Rate vs. Estimated Employed')
plt.xlabel('Labour Participation Rate')
plt.ylabel('Estimated employed')

plt.show()
```
